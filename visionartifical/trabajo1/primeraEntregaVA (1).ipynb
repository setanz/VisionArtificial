## a)  A continuaci√≥n, se presenta una transformaci√≥n sobre ambas im√°genes 

<br> <center> $ G(x,y) = (1-k)*I1(x,y) + k*I2(x,y) $ </center> </br>

<br> <center> $ 0 < k < 1 $ </center> </br>

<ul> Realice un c√≥digo que muestre esta transformaci√≥n, posteriormente responda: </ul>

<ul> <br> ¬øQu√© ocurre si k = 1, k = 0, k = 0.5? </br> </ul>

```python
#Se multiplican los valores de cada canal de las imagenes por (1-K) unidades, con 0 < K < 1
#Para k =1

k = 0.5
p = (1-k)

#Se crea la imagen copia
img1_div = np.zeros(img1.shape, np.uint8)
img2_div = np.zeros(img2.shape, np.uint8)

#Se sustituyen los valores de cada canal de las copias por los valores modificados de la imagen original

#se multiplca por (1-k) la imagen 1
img1_div[:,:,0] = cv2.multiply(img1[:,:,0], p)
img1_div[:,:,1] = cv2.multiply(img1[:,:,1], p)
img1_div[:,:,2] = cv2.multiply(img1[:,:,2], p)


#se multiplca por k la imagen 2
img2_div[:,:,0] = cv2.multiply(img2[:,:,0], k)
img2_div[:,:,1] = cv2.multiply(img2[:,:,1], k)
img2_div[:,:,2] = cv2.multiply(img2[:,:,2], k)

#hacerlas del mismo tama√±o para poderlas sumar

img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

#Indicamos el par√°metro a que indica la importancia de cada imagen


#Mediante la T.Lineal, si a=1, tenemos la imagen 1 √∫nicamente. Si a=0, obtenemos solo img2.
# Si 0<a<1 tendremos combinaciones ponderadas de las im√°genes entrada
res = cv2.add(img2*k,img1*(1-k)).astype(np.uint8)


#Se realiza el arreglo de 2x2 para mostrar imagen modificada vs histogramas
fig, arreglo_plots = plt.subplots(1,4, figsize=(15, 3))
colors = ('r','g','b')

#Se le dan t√≠tulos a los cuadros, se imprimen la img 1 y su histograma
arreglo_plots[0].set_title('Imagen original')
arreglo_plots[0].imshow(img1, cmap="gray")
arreglo_plots[1].set_title('Histograma original')
for i,col in enumerate(colors):
    img_array_i = img1[:,:,i].ravel()
    arreglo_plots[1].hist(img_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[i])

arreglo_plots[2].set_title('Imagen modificada k=1')
arreglo_plots[2].imshow(img1_div, cmap="gray")
arreglo_plots[3].set_title('Histograma modificado k=1')
for i,col in enumerate(colors):
    img_array_i = img1_div[:,:,i].ravel()
    arreglo_plots[3].hist(img_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[i])
    
fig,arreglo_plots2 = plt.subplots(1,4, figsize=(15, 3))
    
#Se le dan t√≠tulos a los cuadros, se imprimen la img2 y su histograma
arreglo_plots2[0].set_title('Imagen original')
arreglo_plots2[0].imshow(img2, cmap="gray")
arreglo_plots2[1].set_title('Histograma original')
for i,col in enumerate(colors):
    img_array_i = img2[:,:,i].ravel()
    arreglo_plots2[1].hist(img_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[i])

arreglo_plots2[2].set_title('Imagen modificada k=1')
arreglo_plots2[2].imshow(img2_div, cmap="gray")
arreglo_plots2[3].set_title('Histograma modificado k=1')
for i,col in enumerate(colors):
    img_array_i = img2_div[:,:,i].ravel()
    arreglo_plots2[3].hist(img_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[i])    
    
fig,arreglo_plots3 = plt.subplots(1,2, figsize=(15, 3))

#Se le dan t√≠tulos a los cuadros, se imprimen la suma y su histograma
arreglo_plots3[0].set_title('Imagen suma')
arreglo_plots3[0].imshow(res, cmap="gray")
arreglo_plots3[1].set_title('Histograma suma')
for i,col in enumerate(colors):
    img_array_i = res[:,:,i].ravel()
    arreglo_plots3[1].hist(img_array_i,histtype='step', bins=255, 
                            range=(0.0, 255.0),normed=True, color=colors[i])    
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt


img = cv2.imread("res/gato.jpg",cv2.IMREAD_COLOR) #Importa la imagen de gato.jpg que se encuentra en la carpeta res
img_cat = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #convierte la imagen de BGR a RGB

borde = cv2.Canny(img_cat,150,200) #funci√≥n que realiza los bordes

"""

borde = cv2.Canny(img_cat, umbral_minimo, umbral_maximo)

borde: es la imagen resultante despues de aplicar la funci√≥n .Canny, Aparecer√°n los bordes detectados tras el proceso.
img_cat: es la imagen original.
umbral_minimo: es el umbral m√≠nimo en la umbralizaci√≥n por hist√©resis
umbral_maximo: es el umbral m√°ximo en la umbralizaci√≥n por hist√©resis

Pueden darse 3 casos:

*Si el valor del p√≠xel es mayor que el umbral m√°ximo, el p√≠xel se considera parte del borde.
*Un p√≠xel se considera que no es borde si su valor es menor que el umbral m√≠nimo,
*Si est√° entre el m√°ximo y el m√≠nimo, ser√° parte del borde si est√° conectado con un p√≠xel que forma ya parte del borde.

"""

#los contornos quedan en una imagen de escala de grises donde el borde es blanco y el fondo es negro
#Para poder darle color al borde, se necesita pasar la imagen a RGB

borde = cv2.cvtColor(borde, cv2.COLOR_GRAY2RGB) 

#Se define un color para los bordes  (en este caso se escoge el color rojo, para ello se ponen los canales G y B en 0)
#Nota: el borde por defecto se da en color blanco el cual posee los canales RGB en un valor de 255 cada uno

borde[:,:,1]=0 
borde[:,:,2]=0

#Se define una figura para la impresi√≥n de la imagen original, la imagen con los bordes y la suma de ambas imagenes

fig = plt.figure(figsize=(20,30))

plt.subplot(1,3,1)
plt.imshow(img_cat,cmap='gray')
plt.title("Imagen Original")

plt.subplot(1,3,2)
plt.imshow(borde,cmap='gray')
plt.title("Bordes extraidos")

plt.subplot(1,3,3)
plt.imshow(cv2.add(img_cat,borde),cmap='gray')
plt.title("Bordes resaltados en la imagen original");
```
